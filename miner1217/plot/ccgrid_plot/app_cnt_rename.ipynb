```python
import pickle
import numpy
import math
import matplotlib
from matplotlib.colors import LogNorm
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import FuncFormatter
import numpy as np
from collections import OrderedDict

def plotTopIOApp(x_list, y_list, z_list):
    plt.gcf().clear()
 #   plt.xlabel('Applications', fontsize = 18)
    plt.ylabel('Job Count', fontsize = 18)
    title = "app_cnt"
    cursor = 0

    x = range(0, 15)
    print ("ylist is \n")
    print (y_list)
    print ("xlist is \n")
    print (x_list)
    plt.xticks(x, x_list)
    rects = plt.bar(x, y_list, color = 'b')
#    plt.yscale('log')
    plt.xticks(x, x_list, fontsize=15, rotation = 90)
    plt.yticks(fontsize = 18)
    plt.legend()
    for rect, label in zip(rects, z_list):
        height = rect.get_height()
        #print "x is %d, label:%s"%(rect.get_x(), label)
        plt.text(rect.get_x() + rect.get_width() / 2, height + 1, label,
                ha='center', va='bottom', fontsize = 8, color = 'red')
 #   plt.show()
    plt.savefig(title+'.jpg', format = 'jpg', bbox_inches='tight')
    
def get_name(name):
    if name == "cesm.exe":
        return "Climate1"
    if "Nyx" in name:
        return "Cosmology2"
    if name == "hacc_tpm":
        return "Cosmology1"
    if name == "pw.x":
        return "Quantum1"
    if name == "s3d.x":
        return "Combustion1"
    if name == "vasp_std":
        return "Quantum2"
    if name == "elm_6f":
        return "Physics1"
    if name == "wrf.exe":
        return "Climate2"
    if name == "phoenix":
        return "Astro1"
    if name == "cgyro":
        return "Fusion1"
    if name == "gene_cori":
        return "Biology1"
    if "sgw.x" in name:
        return "Material2"
    if "pot" in name:
        return "Material4"
    if "delay" in name:
        return "Material1"
    if "tc.x" in name:
        return "Material3"
    return ""
    
    
        
tmp_str = "/global/cscratch1/sd/tengwang/miner1217/app_cnt.log"
save_fd = open(tmp_str, 'rb')
app_dict = pickle.load(save_fd)
save_fd.close()

sorted_dict = sorted(app_dict.items(), key = lambda x:-x[1][0])
total = 0
cycle_total = 0
for record in sorted_dict:
 #   print "key:%s, value:%d\n"%(record[0], record[1])
    total += record[1][1]
    cycle_total += record[1][0]
    
x_lst = []
y_lst = []
z_lst = []
count = 0
cycle_count = 0
cursor = 0
for record in sorted_dict:
    count += record[1][1]
    name = get_name(record[0])
    if name == "":
        name = record[0]

    x_lst.append(name)
    y_lst.append(int(record[1][1]))
    z_lst.append(str(round(float(record[1][0])*100/cycle_total, 1))+"%")
    cycle_count += float(record[1][0])
   # if name == "Nyx3d":
    print ("app:%s, count:%d, percent:%lf, tpercent:%lf, cycle:%lf\n"%(name, record[1][1], float(record[1][0])/cycle_total *100, float(count)/total * 100,float(record[1][0])))
    cursor += 1
    if cursor == 15:
        print ("count:%lf"%(float(count)/total*100))
        print ("cycle_count:%lf"%(float(cycle_count)/cycle_total))
        break
   # if float(count)/total*100 > 90:
   #     break
print ("cursor is %d\n"%cursor)
print ("tot_count is %d\n"%count)
print ("tot_cycles is %ld\n"%cycle_count)
plotTopIOApp(x_lst, y_lst, z_lst)

    
```
