```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pickle

def plotTopIOApp(app_list, io_type):
    plt.gcf().clear()
 #   plt.xlabel('Applications', fontsize = 18)
    plt.ylabel('Data Size (TB)', fontsize = 18)
    x = []
    y = []
    xlabels = []
    title = "top_app_%s"%io_type
    cursor = 0
    if io_type == "READ":
        x_labels = ["Chemistry1", "Chemistry2", "Computer Science", "Chemistry3", "Materials", "Cosmology", "App1", "App2", "App3", "Physics1"]
    if io_type == "WRITE":
        x_labels = ["Computer Science", "Cosmology", "Chemistry1", "Cosmology", "App1", "Physics1", "Physics2", "Physics3", "Chemistry", "Physics4"] 
    for record in app_list:
        x.append(cursor)
      #  if "qcprog" in record[0]:
        #    xlabels.append("qcprog")
       #     xlabels.append(record[0])
       # else:
        #    if "absorption" in record[0]:
         #       xlabels.append(record[0])
          #  else:
           #     xlabels.append(record[0])
#        xlabels.append(tmp_key)
        y.append(record[1]/1024/1048576/1024)
        cursor = cursor + 1
        if cursor == 10:
            break;
    rects = plt.bar(x, y, color = 'b')
#    plt.yscale('log')
    plt.xticks(x, x_labels, fontsize=15, rotation = 90)
    plt.yticks(fontsize = 18)
    plt.legend()
    
#    for rect, label in zip(rects, y_labels):
#        height = rect.get_height()
        #print "x is %d, label:%s"%(rect.get_x(), label)
#        ax.text(rect.get_x() + rect.get_width() / 2, height + 1, label,
#                ha='center', va='bottom', fontsize = 14, color = 'red')
    plt.show()
 #   plt.savefig(title+'.pdf', format = 'pdf', bbox_inches='tight')
    
plot_path = "/global/cscratch1/sd/tengwang/miner1217/top_read_app.pkl"
save_fd = open(plot_path, 'rb')
app_list = pickle.load(save_fd)
plotTopIOApp(app_list, "READ")
```
