```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pickle

plot_path = "/global/cscratch1/sd/tengwang/miner1217/read_write_ratio.pkl"

save_fd = open(plot_path, 'rb')
y_list = pickle.load(save_fd)
save_fd.close()

cur_sum = 0
for i in range(0, len(y_list)):
    cur_sum += y_list[i]
    
sub_sum = 0
for i in range(90,100):
    sub_sum += y_list[i]
    
print "write only:%lf\n"%(float(y_list[0])/cur_sum)
print "read only:%lf\n"%(float(y_list[99])/cur_sum)
print "90th is %lf\n"%(float(sub_sum)/cur_sum)

x_list = range(1,101)
plt.gcf().clear()
fig, ax = plt.subplots()
rects = ax.bar(x_list, y_list, color = 'b')

ax.set_xticks([0, 20, 40, 60, 80, 100])
ax.set_xticklabels([0, 20, 40, 60, 80, 100], fontsize = 16)
ax.set_yticks((0, 20000, 40000, 60000, 80000, 100000))
ax.set_yticklabels((0, 20000, 40000, 60000, 80000, 100000), fontsize = 16)

ax.set_xlabel('Read Percent', fontsize = 18)
ax.set_ylabel('Job Count', fontsize = 18)
plt.show()
#title="read_write_ratio"
#plt.savefig(title+'.pdf', format = 'pdf', bbox_inches='tight')
```
